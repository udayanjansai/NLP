**Stemming**

Stemming reduces words to their root form.

Stemming is an important text-processing technique that reduces words to their base or root form by removing prefixes and suffixes.
This process standardizes words which helps to improve the efficiency and effectiveness of various natural language processing (NLP)
tasks.

**Examples of stemming for the word "like":**
"likes" - "like"

"liked" - "like"

"likely" - "like"

```python
from nltk.stem import PorterStemmer
porter_stemmer=PorterStemmer()
words=['likes','playing','eating','happily','doing']
stemmed_words=[porter_stemmer.stem(word) for word in words]
print('original: ',words)
print('stemmed: ',stemmed_words)
```

Lemmatization produces meaningful root words.

**Example**

Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base
form known as a "lemma." For example, the lemma of "running" is "run" and "better" becomes "good."

**Used for:**

Reduced Data Redundancy

Better NLP Model Performance

**Text Representation Techniques**

Machines cannot understand text.

They understand numbers.

So we convert text - numbers.

**Bag of Words (BoW)**

BoW represents text as word frequency vectors.
In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can
understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a
collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the
words or their grammar but focuses on counting how often each word appears in the text.
This makes it useful for tasks like text classification, sentiment analysis and clustering.

**Step 1: Preprocessing the Text**

Clean the text

Convert to lowercase

Remove special characters

```python
import re
text=[
    'I love Coding!',
    'I love python#$%',
    'Python Coding is fun'
]
dataset=[]
for sentence in text:
  sentence=sentence.lower()
  sentence=re.sub(r'[^\w\s]','',sentence)
  dataset.append(sentence)
print(dataset)
```

```python
from collections import defaultdict
wordcount=defaultdict(int)
for sentence in dataset:
  for word in sentence.split():
    wordcount[word]+=1
print(wordcount)
```

**Step 3: Selecting Most Frequent Words**

Let's take top 3 words

```python
import heapq
freq_words=heapq.nlargest(3,wordcount,key=wordcount.get)
print(freq_words)
```

**Step 4: Build Bag of Words Matrix**

Now we create a matrix:

Rows = sentences

Columns = frequent words

Value = 1 (present) or O (absent)

```python
x=[]
for sentence in dataset:
  vector=[]
  words=sentence.split()
  for word in freq_words:
    if word in words:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```

**N-grams**

N-grams capture word sequences.

N-gram is a technique that splits text into sequences of N consecutive words to capture context and word order for machine learning
models.

N-gram = group of N consecutive words
1 word - Unigram
2 words - Bigram
3 words - Trigram

**Key Idea**
dr N-grams capture word relationships
BoW - "love", "coding"
Bigram - "love coding" V (more meaningful)

**Simple Example**

Sentence:

"I love coding"

Unigrams (N = 1)
["l", "love", "coding"]

· Bigrams (N = 2)
["I love", "love coding"]

Trigrams (N = 3)
["I love coding"]

```python
import re
text=[
    'I love Coding!',
    'I love python#$%',
    'Python Coding is   fun'
]
dataset=[]
for sentence in text:
  sentence=sentence.lower()
  sentence=re.sub(r'[^\w\s]','',sentence)
  sentence=re.sub(r'\s+',' ',sentence)
  dataset.append(sentence)
print(dataset)
```

```python
def generate_ngrams(sentence,n):
  words=sentence.split()
  ngrams=[]
  for i in range(len(words)-n+1):
    ngrams.append(' '.join(words[i:i+n]))
  return ngrams
```

```python
bigrams=[]
trigrams=[]
for sentence in dataset:
  bigram=generate_ngrams(sentence,2)
  bigrams.extend(bigram)
print(bigrams)
for sentence in dataset:
  trigram=generate_ngrams(sentence,3)
  trigrams.extend(trigram)
print(trigrams)
```

```python
ngram2count=defaultdict(int)
for ngram in bigrams:
  ngram2count[ngram]+=1
print(ngram2count)
```

```python
freq_bigrams=heapq.nlargest(3,ngram2count,key=ngram2count.get)
print(freq_bigrams)
```

```python
# Build Matrix
X = []

for sentence in dataset:
  vector = []
  sentence_ngrams = generate_ngrams(sentence, 2)

  for gram in freq_bigrams:
    if gram in sentence_ngrams:
      vector.append(1)
    else:
      vector.append(0)

  X.append(vector)

print(X)

```

Word Embeddings:

Word Embeddings = representing words as meaningful numbers (vectors)

Unlike BoW/TF-IDF:

They don't just count words X

They capture meaning and relationships

Why Do We Need Word Embeddings?

Problem with TF-IDF / BoW
Example:

"good" and "excellent"

BoW / TF-IDF:

Treat them as completely different X


Word Embeddings Fix This

They understand:

"good" = "excellent"

"king" - "man" + "woman" = "queen"

Key Idea

Each word is converted into a vector (list of numbers)

Example:

"king" - [O.2, 0.8, 0.1]

"queen" - [O.25, 0.75, 0.15]

Similar words -> similar vectors

Types of Word Embeddings

1. Word2Vec :
Word2Vec is a technique to create word embeddings (vectors) by learning from context.

Idea:

"Words that appear in similar contexts have similar meanings"

Two methods:

CBOW (predict word from context)

Skip-gram (predict context from word)

1. CBOW (Continuous Bag of Words)

Idea:

Predict the center word using surrounding words

Example

Sentence:

"I love coding"

Input (context words):

"l"and "coding"

Output (target word):

"love"

2. Skip-gram

Idea:

Predict surrounding words using center word

Example

Sentence:
"I love coding"

Input (center word):
"love'

Output (context words):
"l", "coding"

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string

def preprocess_text(text):
    # 1. Lowercasing
    text = text.lower()

    # 2. Tokenization
    tokens = word_tokenize(text)

    # Initialize lemmatizer and stopwords
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))

    processed_tokens = []
    for token in tokens:
        # 3. Punctuation removal (and filter out empty strings)
        token = token.translate(str.maketrans('', '', string.punctuation))
        if not token.isalpha():  # Ensure token is purely alphabetic after punctuation removal
            continue

        # 4. Stopword removal
        if token not in stop_words:
            # 5. Lemmatization
            # Lemmatize with a default POS tag 'n' (noun) if not specified for better results.
            # For more advanced lemmatization, you'd perform POS tagging first.
            processed_tokens.append(lemmatizer.lemmatize(token))

    return processed_tokens

# Example usage:
sample_text = "This is an example sentence to test the preprocessing function. It contains some stopwords and punctuation! Running it should yield clean, base words."
cleaned_tokens = preprocess_text(sample_text)
print(cleaned_tokens)
```

```python
df['processed_text'] = df['text'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Join the list of processed tokens back into a string for vectorization
df['processed_text_str'] = df['processed_text'].apply(lambda x: ' '.join(x))

# CountVectorizer (BoW)
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(df['processed_text_str'])

# TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['processed_text_str'])

print("CountVectorizer - Feature size (number of unique words):", X_count.shape[1])
print("TfidfVectorizer - Feature size (number of unique words):", X_tfidf.shape[1])

print("\nCountVectorizer - Sparsity (percentage of zero values):", (X_count.nnz / (X_count.shape[0] * X_count.shape[1])) * 100)
print("TfidfVectorizer - Sparsity (percentage of zero values):", (X_tfidf.nnz / (X_tfidf.shape[0] * X_tfidf.shape[1])) * 100)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create a dictionary to store model results
results = {}

def add_model_results(model_name, y_true, y_pred, results_dict):
    accuracy = accuracy_score(y_true, y_pred)
    # For binary classification, 'pos_label' needs to be specified for precision, recall, f1-score
    # Assuming 'spam' is the positive class
    precision = precision_score(y_true, y_pred, pos_label='spam')
    recall = recall_score(y_true, y_pred, pos_label='spam')
    f1 = f1_score(y_true, y_pred, pos_label='spam')
    results_dict[model_name] = {'Accuracy': accuracy, 'Precision (Spam)': precision, 'Recall (Spam)': recall, 'F1-Score (Spam)': f1}

# Add results for already trained models (from previous cells)
add_model_results('Naive Bayes (CountVectorizer)', y_test, y_pred_count, results)
add_model_results('Naive Bayes (TfidfVectorizer)', y_test, y_pred_tfidf, results)
add_model_results('Logistic Regression (TfidfVectorizer)', y_test, y_pred_log_reg, results)

# Add results for newly trained models
add_model_results('Logistic Regression (CountVectorizer)', y_test, y_pred_log_reg_count, results)
add_model_results('SVM (CountVectorizer)', y_test, y_pred_svm_count, results)
add_model_results('SVM (TfidfVectorizer)', y_test, y_pred_svm_tfidf, results)

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results).T # Transpose to have models as rows
display(results_df.round(4))
```

### Convert Sentences to Vectors (Average Embeddings)

```python
import numpy as np

def sentence_vector(word_list, model, vector_size):
    # Filter out words not in the model's vocabulary
    filtered_words = [word for word in word_list if word in model.wv]
    if not filtered_words:
        return np.zeros(vector_size)

    # Average the word vectors
    return np.mean(model.wv[filtered_words], axis=0)

# Apply the function to create sentence vectors
df['sentence_vector'] = df['processed_text'].apply(lambda x: sentence_vector(x, word2vec_model, 100))

# Convert the list of arrays to a 2D numpy array for model training
X_word2vec = np.vstack(df['sentence_vector'].values)

print(f"Shape of sentence vectors (X_word2vec): {X_word2vec.shape}")
print("\nFirst 5 sentence vectors (partial view):\n", X_word2vec[:5, :5]) # Display first 5 vectors, first 5 dimensions
```

```python
from collections import defaultdict, Counter
import nltk

# Filter for spam messages (This line was added to fix the NameError)
spam_df = df[df['message'] == 'spam']

# Flatten the list of processed spam messages into a single list of words
all_spam_words = [word for sublist in spam_df['processed_text'] for word in sublist]

# Generate bigrams
bigrams = list(nltk.bigrams(all_spam_words))

# Count bigram frequencies
bigram_counts = Counter(bigrams)

# Create a dictionary to store word counts for unigrams (needed for probability calculation)
unigram_counts = Counter(all_spam_words)

# Function to predict the next word
def predict_next_word(current_word, num_predictions=1):
    current_word = current_word.lower() # Ensure case-insensitivity
    possible_next_words = defaultdict(float)

    # Calculate probability for each possible next word
    for (word1, word2), count in bigram_counts.items():
        if word1 == current_word:
            # P(word2|word1) = Count(word1, word2) / Count(word1)
            possible_next_words[word2] = count / unigram_counts[current_word]

    if not possible_next_words:
        return [("No prediction", 0.0)]

    # Sort predictions by probability (descending)
    sorted_predictions = sorted(possible_next_words.items(), key=lambda item: item[1], reverse=True)

    return sorted_predictions[:num_predictions]

# Example Usage:
print("\n--- Bigram Next Word Prediction ---")
example_word = "win"
predictions = predict_next_word(example_word, num_predictions=3)
print(f"Given '{example_word}', the most likely next words are:")
for word, prob in predictions:
    print(f"- '{word}' (Probability: {prob:.4f})")

example_word = "money"
predictions = predict_next_word(example_word, num_predictions=3)
print(f"\nGiven '{example_word}', the most likely next words are:")
for word, prob in predictions:
    print(f"- '{word}' (Probability: {prob:.4f})")

example_word = "free"
predictions = predict_next_word(example_word, num_predictions=3)
print(f"\nGiven '{example_word}', the most likely next words are:")
for word, prob in predictions:
    print(f"- '{word}' (Probability: {prob:.4f})")
```
